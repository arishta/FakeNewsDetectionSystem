### STEP 2: PROCESS/PREPARE THE DATA
Pre-processing data involves the following steps:

####  Remove the NULL values 
We replace the NULL values with blank space - ' '.

#### Create a new attribute
We have created a new attribute - *total* which is a combination of *title* ,  *author*   and   *text* in both the training and testing dataset. This allow all the relevant information to be contained in just one attribute which will help simplify things. When training our model, we can use the new attribute *total* instead of the other three attributes.

#### Convert to lowercase
We convert all the text data to lowercase.

#### Remove non ASCII values
We remove the non ASCII values from the text data.

#### Remove punctution
We remove all the punctuations from the text data.

#### Removing stopwords
Stopwords are some common occurring words in the English language which do not add much value to the text. We remove these words so that they do not take up space in our database or valuable processing time. Hence, we remove these stopwords before we build the model.

#### Tokenization
Tokenization involves splitting sentences and words from the body of the text.

#### Lemmatization
It reduces the words to their root form in order to remove ambiguity from the text.

#### Feature extraction
After tokenization, the words so obtained are encoded as integers or floating point values. This process is called feature extraction which is achieved using *CountVectorizer* and *TfidfVectorizer*.

CountVectorizer is used to transform a corpora of text to a vector of term / token counts.

TfidfVectorizer transforms text to feature vectors that can be used as input to estimators.


```python
#display the number of records that have null values for each of the attributes in the training dataset
print(train.isnull().sum())

print('---------------')

#display the number of records that have null values for each of the attributes in the testing dataset
print(test.isnull().sum())
```

```python
#replace the null values with blank space - ' '
test=test.fillna(' ')
train=train.fillna(' ')
```

```python
#print the first few records of the dataset
train.head()
```

```python
#combine the three attributes - 'title', 'author' and 'text' into a new attribute 'total'
test['total']=test['title']+' '+test['author']+test['text']
train['total']=train['title']+' '+train['author']+train['text']
```

```python
train.head()
```

```python
def removePunctuation(x):
    #Convert all the words to their lowercase form
    x = x.lower()
    #Remove non ASCII chars using RegEx
    x = re.sub(r'[^\x00-\x7f]',r' ',x)
    #Replace all the punctuations with empty spaces using RegEx
    return re.sub("["+string.punctuation+"]", " ", x)
```

```python
import string
train['total'] = train['total'].apply(removePunctuation)
```

```python
#print the value of the 'total' attribute for the first record
train.head(1)['total'][0]
```

```python
from nltk.corpus import stopwords
```

```python
#Remove stopwords from the text
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')

train['total'] = train['total'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
train.head(1)['total'][0]
```

```python
#Tokenization to break the text into smaller chunks and lemmatization to reduce the words into their dictionary/root forms
nltk.download('wordnet')
nltk.download('punkt')
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
for index,row in train.iterrows():
    filter_sentence = ''
    sentence = row['total']
    words = nltk.word_tokenize(sentence) #tokenization
    for word in words:
        filter_sentence = filter_sentence + ' ' + str(lemmatizer.lemmatize(word)).lower() #lemmatization
    train.loc[index,'total'] = filter_sentence
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
X_train = train['total']
Y_train = train['label']
```

```python
#Feature extraction using count vectorization and tfidf.
count_vectorizer = CountVectorizer()
count_vectorizer.fit_transform(X_train)
freq_term_matrix = count_vectorizer.transform(X_train)
tfidf = TfidfTransformer(norm="l2")
tfidf.fit(freq_term_matrix)
tf_idf_matrix = tfidf.fit_transform(freq_term_matrix)
```

```python
tf_idf_matrix
```

```python
test_counts = count_vectorizer.transform(test['total'].values)
test_tfidf = tfidf.transform(test_counts)
```

### STEP 3: BASIC DATA EXPLORATION

#### Count plot
A count plot is used to show the counts of observations in each categorical bin using bars. Here, we have plotted one between number of records and label(real/fake).

#### WordCloud
Word Cloud is a data visualization technique used for representing text data in which the size of each word indicates its frequency or importance.

```python
#Count plot between number of records and label(real or fake)
#If label is 1, then the news item is real, else fake

print(train.groupby(['label'])['total'].count())
train.groupby(['label'])['total'].count().plot(kind="bar")
plt.show()
```

```python
real_words = ''
fake_words = ''
stopwords = set(STOPWORDS) 

#We combine all the text which belongs to real news items in the variable real_words
for val in train[train['label']==1].total: 
    # split the value 
    tokens = val.split() 
    real_words += " ".join(tokens)+" "
    
#We combine all the text which belongs to real news items in the variable fake_words
for val in train[train['label']==0].total:   
    # split the value 
    tokens = val.split() 
    fake_words += " ".join(tokens)+" "
```

```python
#WordCloud for real_words
#All the stopwords are excluded from the WordCloud
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(real_words) 
  
# plot the WordCloud                       
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

```python
#WordCloud for fake_words
#All the stopwords are excluded from the WordCloud
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='black', 
                stopwords = stopwords, 
                min_font_size = 10).generate(fake_words) 
  
# plot the WordCloud                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

### STEP 4: TRAIN AND TEST THE MODEL

#### Train the model
We train our model using logistic regression. 
Why logistic regression?
Logistic regression is a classification algorithm, used when the value of the target variable is categorical in nature and when the data in question has binary output, i.e., when it belongs to one class or another, or is either a 0 or 1.

#### Test the model
After training the model, we test it using the testing dataset. Accuracy and confusion matrix will give us a picture of how well our model has perfomed wrt the training dataset.




```python
#We have combined 'title','author' and 'text' into 'total', so we only consider the 'total' as the independent attribute
train = train[['total','label']]
```

```python
#split in samples
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(tf_idf_matrix, Y_train, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score, classification_report, accuracy_score, confusion_matrix
```

```python
#training model using logistic regression
logreg = LogisticRegression(C=1e5)
logreg.fit(X_train, y_train)
pred = logreg.predict(X_test)
#accuracy of training and testing dataset
print('Accuracy of Logistic classifier on training set: {:.2f}'
     .format(logreg.score(X_train, y_train)))
print('Accuracy of Logistic classifier on test set: {:.2f}'
     .format(logreg.score(X_test, y_test)))
cm = confusion_matrix(y_test, pred)
cm
```

### STEP 5: CREATE A PIPELINE

Our model has been trained, and hence we put it into a pipeline, so that we can use it in our web application.

```python
#Assign the variables again as once transformed vectors cannot be transformed again using pipeline
X_train = train['total']
Y_train = train['label']
```

```python
from sklearn.pipeline import Pipeline
import joblib
from sklearn import linear_model
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
pipeline = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer(norm='l2')),
    ('clf', linear_model.LogisticRegression(C=1e5)),
])
```

```python
pipeline.fit(X_train, Y_train)
```

```python
#save the pipeline
filename = 'pipeline.sav'
joblib.dump(pipeline, filename)
```
